```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

edff_erosita = pd.read_csv('data/result/merged_catalog_edff_erosita.csv')
edff_cdfs = pd.read_csv('data/result/merged_catalog_edff_cdfs.csv')
```

```python
# EDFF_EROSITA 
def create_detailed_stats(df, name):
    stats = []
    
    for column in df.columns:
        stats_dict = {
            'Column': column,
            'Data Type': str(df[column].dtype),
            'Mean': df[column].mean() if pd.api.types.is_numeric_dtype(df[column]) else None,
            'Median': df[column].median() if pd.api.types.is_numeric_dtype(df[column]) else None,
            'Std Dev': df[column].std() if pd.api.types.is_numeric_dtype(df[column]) else None
        }
        stats.append(stats_dict)
    
    return pd.DataFrame(stats)

# Create statistics for both DataFrames
erosita_stats = create_detailed_stats(edff_erosita, 'eROSITA')
cdfs_stats = create_detailed_stats(edff_cdfs, 'CDFS')
```

```python
erosita_stats.to_csv('data/result/erosita_stats.csv', index=False)

catalog_matching_stats = erosita_stats[0:5]
edff_columns = erosita_stats[5:(5+130)]
erosita_columns = erosita_stats[(5+130):]
```

```python
import dataframe_image as dfi

def save_df_as_png(df, filename, index=True):
    """
    Save a DataFrame as a styled PNG image
    
    Parameters:
    -----------
    df : pandas DataFrame
        The DataFrame to save
    filename : str
        The output filename (should end with .png)
    index : bool, default=True
        Whether to include the index in the output
    """
    # Create a styled version of the DataFrame
    styled_df = df.style.set_properties(**{
        'text-align': 'left',
        'font-size': '11pt',
        'border-color': 'black',
        'border-style': 'solid',
        'border-width': '1px',
        'padding': '5px'
    })
    
    # Add header styling
    styled_df = styled_df.set_table_styles([
        {'selector': 'th', 
         'props': [('background-color', '#f2f2f2'), 
                   ('color', 'black'),
                   ('font-weight', 'bold'),
                   ('text-align', 'left'),
                   ('border-color', 'black'),
                   ('border-style', 'solid'),
                   ('border-width', '1px'),
                   ('padding', '5px')]}
    ])
    
    # Alternating row colors for better readability
    styled_df = styled_df.set_table_styles([
        {'selector': 'tr:nth-child(even)', 
         'props': [('background-color', '#f9f9f9')]}
    ], overwrite=False)
    
    # Save to PNG
    dfi.export(styled_df, filename, table_conversion='matplotlib')
    print(f"Table saved as {filename}")

# Save the statistics tables as PNG files
save_df_as_png(erosita_stats, 'images/erosita_stats.png')
save_df_as_png(cdfs_stats, 'images/cdfs_stats.png')

# Display confirmation
print("Both tables have been saved as PNG files.")
```
