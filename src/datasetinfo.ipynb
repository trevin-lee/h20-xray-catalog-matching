# Dataset Column Headers

```python
from tabulate import tabulate

max_len = max(len(df_H20.columns), len(df_eROSITA.columns), len(df_CDFS.columns))
columns_H20 = list(df_H20.columns) + [''] * (max_len - len(df_H20.columns))
columns_eROSITA = list(df_eROSITA.columns) + [''] * (max_len - len(df_eROSITA.columns))
columns_CDFS = list(df_CDFS.columns) + [''] * (max_len - len(df_CDFS.columns))


table = [(i, columns_H20[i], columns_eROSITA[i], columns_CDFS[i]) for i in range(max_len)]

print(tabulate(table, headers=['Index', 'df_H20_columns', 'df_eROSITA_columns', 'df_CDFS_columns'], tablefmt='grid'))
```
